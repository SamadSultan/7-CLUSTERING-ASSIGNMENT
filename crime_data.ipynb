```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('crime_data.csv')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data
```

```python
data2=StandardScaler().fit_transform(data)
```

```python
data2
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
```

```python
dbscan.fit(data2)
```

```python
dbscan.labels_
```

```python
data['clusters']=dbscan.labels_
```

```python
data
```

```python
data.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(7, 7))  
plt.scatter(data['clusters'],data['Murder'], c=dbscan.labels_);
```

```python
plt.figure(figsize=(7, 7))  
plt.scatter(data['clusters'],data['Assault'], c=dbscan.labels_);
```

```python
plt.figure(figsize=(7, 7))  
plt.scatter(data['clusters'],data['UrbanPop'], c=dbscan.labels_);
```

```python
plt.figure(figsize=(7, 7))  
plt.scatter(data['clusters'],data['Rape'], c=dbscan.labels_);
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(data['clusters'],data['Rape'], c=kmeans.labels_)
```

```python
plt.figure(figsize=(10, 7))  
dendogram=sch.dendrogram(sch.linkage(data,'complete'))
```
